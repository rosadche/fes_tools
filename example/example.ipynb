# 1D Example

## Load our data

```python
example_dfs_1d = [utils.read_plumed_fes(f"./data/1d/dirpt_asp_walks_24_fes_hills_{i}.dat") for i in range(30000, 288000 + 2000, 2000) ]
```

## This example shows how automatic guessing of the product location can go wrong (static graph)

```python
example_1d_fes = fes.fes(example_dfs_1d, cv_cols=["pt_assym_stretch"], fe_col="file.free")
```

```python
options = dict()
options["cv_labels"]       = [r"Proton Transfer Assymetric Stretch ($\AA$)"]
options["time_unit"]       = r"Hills"
options["time_vals"]           = list(range(30000, 288000 + 2000, 2000))
options["title"]           = "DFTB Metadynamics Example: 1D Surface"
options["energy_unit_str"] = r"$\mathrm{kcal \cdot mol^{-1}}$"
options["track"]           = False
fig_1d, axs_1d = plotting.plot_fes(example_1d_fes, options)
```

## This example shows how it can be fixed by supplying the correct general region

```python
example_1d_fes = fes.fes(example_dfs_1d, cv_cols=["pt_assym_stretch"], fe_col="file.free", product_loc=[[2], [7]])
```

```python
fig_1d, axs_1d = plotting.plot_fes(example_1d_fes, options)
```

## Example showing tracking over time but without making an animation

```python
options = dict()
options["cv_labels"]       = [r"Proton Transfer Assymetric Stretch ($\AA$)"]
options["time_unit"]       = r"Hills"
options["time_vals"]           = list(range(30000, 288000 + 2000, 2000))
options["title"]           = "DFTB Metadynamics Example: 1D Surface"
options["energy_unit_str"] = r"$\mathrm{kcal \cdot mol^{-1}}$"
options["track"]           = True
options["gif"]             = False
fig_1d_track, axs_1d_track = plotting.plot_fes(example_1d_fes, options)
```

## Example showing tracking over time but this time with making an animation, and saving a GIF and PNG with the "save_file" file handle

```python
options = dict()
options["cv_labels"]       = [r"Proton Transfer Assymetric Stretch ($\AA$)"]
options["time_unit"]       = r"Hills"
options["time_vals"]           = list(range(30000, 288000 + 2000, 2000))
options["title"]           = "DFTB Metadynamics Example: 1D Surface"
options["energy_unit_str"] = r"$\mathrm{kcal \cdot mol^{-1}}$"
options["track"]           = True
options["gif"]             = True
options["save_file"]       = "test"
fig_1d_track, axs_1d_track = plotting.plot_fes(example_1d_fes, options)
```

# 2D Example

```python
# utils, plotting, fes
from fes_tools import *
example_dfs_2d = [utils.read_plumed_fes(f"./data/2d/bias_hemi1_prop_{i}.dat") for i in range(510000, 526000 + 4000, 4000) ]
```

```python
example_2d_fes = fes.fes(example_dfs_2d, cv_cols=["nucattack", "zeta"], fe_col="fes", product_loc=[[0, 0.25], [1, 1]])
```

## Example with tracking, no gif

```python
options = dict()
options["cv_labels"]       = [r"Nucleophilic Attack Assymetric Stretch ($\AA$)", "Proton Transfer ($\zeta$)"]
options["time_unit"]       = r"Hills"
options["time_vals"]           = list(range(510000, 526000 + 4000, 4000))
options["title"]           = "DFTB Metadynamics Example: 2D Surface"
options["energy_unit_str"] = r"$\mathrm{kcal \cdot mol^{-1}}$"
options["track"]           = True
options["gif"]             = False

fig_2d, ax_2d = plotting.plot_fes(example_2d_fes, options)
```

## Example with tracking and saving an animation

```python
options = dict()
options["cv_labels"]       = [r"Nucleophilic Attack Assymetric Stretch ($\AA$)", "Proton Transfer ($\zeta$)"]
options["time_unit"]       = r"Hills"
options["time_vals"]           = list(range(510000, 526000 + 4000, 4000))
options["title"]           = "DFTB Metadynamics Example: 2D Surface"
options["energy_unit_str"] = r"$\mathrm{kcal \cdot mol^{-1}}$"
options["track"]           = True
options["gif"]             = True
options["save_file"]      = "test_2d"

fig_2d, ax_2d = plotting.plot_fes(example_2d_fes, options)
```

## Example with tracking and saving an animation, using NEB algorithm

```python
options = dict()
options["cv_labels"]       = [r"Nucleophilic Attack Assymetric Stretch ($\AA$)", "Proton Transfer ($\zeta$)"]
options["time_unit"]       = r"Hills"
options["time_vals"]           = list(range(510000, 526000 + 4000, 4000))
options["title"]           = "DFTB Metadynamics Example: 2D Surface"
options["energy_unit_str"] = r"$\mathrm{kcal \cdot mol^{-1}}$"
options["track"]           = True
options["gif"]             = True
options["save_file"]      = "neb_test"

example_2d_fes_neb = fes.fes(example_dfs_2d, cv_cols=["nucattack", "zeta"], fe_col="fes", mfep_type="neb", product_loc=[[0, 0.25], [1, 1]])
fig_2d_neb, ax_2d_neb = plotting.plot_fes(example_2d_fes_neb, options)
```

## Static 1D projection along string

Notice how the NEB algorithm helps find the intermediate comapred to the simple Dijkstra's graph search (next). It also has fewer points. It still finds the transition state but there is less resolution in the curve.

```python
options = dict()
options["cv_labels"]       = [r"Nucleophilic Attack Assymetric Stretch ($\AA$)", "Proton Transfer ($\zeta$)"]
options["time_unit"]       = r"Hills"
options["title"]           = "DFTB Metadynamics Example: 2D Surface Projected to 1D"
options["time_vals"]           = list(range(510000, 526000 + 4000, 4000))
options["energy_unit_str"] = r"$\mathrm{kcal \cdot mol^{-1}}$"
options["track"]           = False
options["gif"]             = False
options["project_1d"]      = True

fig_1d_project, axs_1d_project = plotting.plot_fes(example_2d_fes_neb, options)
```
